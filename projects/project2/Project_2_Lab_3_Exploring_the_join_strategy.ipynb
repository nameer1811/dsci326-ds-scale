```python
water_quality_path = 'data/MinneMUDAC_raw_files/mces_lakes_1999_2014.txt'
xref_path = 'data/MinneMUDAC_raw_files/Parcel_Lake_Monitoring_Site_Xref.txt'
```

```python
water_quality_data = spark.read.csv(water_quality_path, sep='\t', header=True)
xref_data = spark.read.csv(xref_path, sep='\t', header=True)
```

```python
xref_data.take(5) >>to_pandas
```

```python
water_quality_data.take(5) >> to_pandas
```

```python
xref_data.groupBy('Monit_MAP_CODE1').count().take(5) >> to_pandas
```

```python
water_quality_data.groupBy('DNR_ID_Site_Number').count().take(5) >> to_pandas
```

> Both the tables have many-to-many relationship on the columns `DNR_ID_Site_Number` and `Monit_MAP_CODE1`. I think we should use outer join to check if there are mismatches and see if it is a viable choice since it is a many-to-many relation. Outer join will be a good choice since it will join all the matches and then keep `NA` values for the non-matches, it will be easier for us to figure out which matched and which did not.

```python
xref_id = xref_data.select('Monit_MAP_CODE1').distinct()
```

```python
wq_id = water_quality_data.select('DNR_ID_Site_Number').distinct()
```

```python
(xref_id
 .join(wq_id, on=wq_id.DNR_ID_Site_Number == xref_id.Monit_MAP_CODE1, how='anti')
 .count()
)

# 96 xref id's are not in the water quality id's
```

```python
(wq_id
 .join(xref_id, on=wq_id.DNR_ID_Site_Number == xref_id.Monit_MAP_CODE1, how='anti')
 .count()
)
# 5 water quality id's are not in the xref id's
```

> We are going to filter the rest of the lakes leaving the 5 out and we would be fine without those 5 since they would not matter much.

```python
from pyspark.sql.functions import col,isnan, when, count
joined_xref_water_data = (xref_id
 .join(wq_id, on=wq_id.DNR_ID_Site_Number == xref_id.Monit_MAP_CODE1, how='outer')
)
# The alias names are opposite since we are counting the missing values of the data not joined, it made most sense to me
(joined_xref_water_data
.select(count(when(col('Monit_MAP_CODE1').isNull(), 'Monit_MAP_CODE1')).alias('wq_missing_count'),
        count(when(col('DNR_ID_Site_Number').isNull(), 'DNR_ID_Site_Number')).alias('xref_missing_count'))
.collect()
) >> to_pandas
```

> Joining on floating point numbers is problematic since the floating point numbers could be different for different files as they are approximations. They are not guaranteed to be exactly the same on all machines, or even on the same machine at different times. We would want to keep the centroid columns as strings when joining tables because it avoids mismatch between the data. For example, if a lat/long value in one dataset is 92.123456 and the other is 92.123456 then keeping it as string will match but converting to floating point number, one of the dataset could have a the value read in as 92.12346, where the number will be shortened and could also possibly be rounded by the machine. 

```python
from glob import glob
from utility import make_data_frame, get_year
```

```python
make_column_df = lambda year, df: (spark.createDataFrame(pd.DataFrame({'columns':df.columns}))
                                        .select('columns', lit(1).alias(year)))
```

```python
# Place your code/thoughts in one or more code/markdown cells, respectively.
data_file_paths = sorted(glob('./data/MinneMUDAC_raw_files/*parcels.txt'))
parcels = [make_data_frame(file_path) for file_path in data_file_paths if get_year(file_path) != '2002' and get_year(file_path) != '2003']
```

```python
parcel_data_unions = reduce(lambda years, data_frame: years.union(data_frame), [data_frame.select("centroid_lat","centroid_long") for data_frame in parcels])
```

```python
parcel_data_unions.groupBy("centroid_lat","centroid_long").count().orderBy(col('count').desc()).take(5) >> to_pandas
```

```python
xref_data.groupBy("centroid_lat","centroid_long").count().take(5) >> to_pandas
```

> The unioned data of parcels represent the many to many connections while the xref dataset just has one to one relationship. I think the best join type would be a left/right join for this datatype. I was running out of time for this assingment so I used anti-join to calculat the mismatches instead of calculating the matches and mismathces.

```python
xref_geo = xref_data.select("centroid_lat","centroid_long").distinct()
parcel_union_geo = parcel_data_unions.select("centroid_lat","centroid_long").distinct()
```

```python
xref_geo.count()
```

```python
parcel_union_geo.count()
```

```python
# Parcel missing
(xref_geo
 .join(parcel_union_geo, on=["centroid_lat", "centroid_long"], how='anti')
 .count()
)
```

```python
# Xref missing
(parcel_union_geo
 .join(xref_geo, on=["centroid_lat", "centroid_long"], how='anti')
 .count()
)
```

> We will exclude the missing Xrefs and the missing Parcel files since the end goal is property price prediction based on water quality.
