<h1><strong>Prepare to Dive In</strong></h1>
<p>Teams receive access to three core data sets for the event and a potential list of questions to consider for each data set. Because we want this event to be as “real world” as possible, the questions will only provide a starting point for your analysis. The primary goal of your analysis is to provide a thorough and insightful investigation of the relationship between various property characteristics and the water quality of lakes within the seven county Twin Cities region.</p>
<p>We’re especially looking for examples of new, creative and innovative analysis to enhance the impact of your findings. Social Data Science is the primary client for this event, however; other organizations have already expressed interest in your findings. The implications of your results may influence many within this region and beyond.</p>
<p><strong>Main Questions to Consider in Analysis</strong></p>
<p>There are three overarching questions that should be considered in your analysis. These questions will require the use of all three data sets provided below. Additional or supplementary data sets may also be used at your discretion to answer these questions.</p>
<ol>
<li>How does property value influence water quality?</li>
<li>What property characteristics influence water quality the most?</li>
<li>Have changes in land use or land development over time positively or adversely impacted water quality?</li>
</ol>
<p>In addition to the questions above, a potential list of questions has been provided below each data set as guidance. Consider these questions to be a diving board from which you can dive into the water data. All participants are expected to use their originality, creativity and ingenuity to provide the best findings, insights and recommendations possible.</p>
<p>Find the data sets below along with more background information. The data can be downloaded as text files or accessed using BigQuery. Helpful tutorials for using BigQuery are available at the bottom of this page.<br/> <a name="one"></a></p>

<h4><strong>Data Set 1: MetroGIS Regional Tax Parcel Dataset</strong></h4>
<ul>
<ul>
<li>Seven Counties</li>
<li>2002–2015</li>
<li>Approximately 1,500,000 properties</li>
<li>Key Attributes:
<ul>
<li>Tax valuations</li>
<li>Dwelling Type</li>
<li>Property Ownership (public, commercial, residential, etc.)</li>
<li>Description of Use</li>
</ul>
</li>
</ul>
</ul>
<p><strong>Potential questions for consideration:</strong></p>
<ul>
<li>Property ownership is an investment and the hope is that your investment will gain value over time. What locations have seen the most substantial increase / decrease in property values over time?</li>
<li>The type of dwelling or type of property owner (i.e. public, commercial, residual, etc.) may adversely impact water quality. What locations have seen the most change or growth in these characteristics over time?</li>
<li>Local and county governments are often responsible for developing policies and regulating land development. What locations have seen the most change in land development over time? What variations exist in how land is being developed across various local and county governments?</li>
</ul>

<h4><strong>Data Set 2:</strong> <strong>Metropolitan Council Lake Monitoring Data</strong></h4>
<ul>
<li>1980–2014</li>
<li>332 Lakes</li>
<li>Key Attributes:
<ul>
<li>Seasonal Lake Grade</li>
<li>Physical Condition</li>
<li>Recreational Suitability</li>
<li>Secchi Depth</li>
<li>Total Phosphorus</li>
</ul>
</li>
</ul>
<p><strong>Potential questions for consideration:</strong></p>
<ul>
<li>What lakes tend to have the best / worst water quality? Are the lakes with the best / worst water quality consistent over time?</li>
<li>What watersheds tend to have the best / worst water quality? Are the watersheds with the best / worst water quality consistent over time?</li>
<li>One may anticipate that the water quality attributes of lakes within a watershed depend on each other. Does this data support this notation? How does time influence this dependence structure?</li>
</ul>

<h4><strong>Data Set 3: Social Data Science Water Proximity Data</strong></h4>
<ul>
<li>Cross reference of Regional Tax Parcel &amp; Lake Monitoring Data sets</li>
<li>Key Attributes:
<ul>
<li>Distance of Tax Parcel to nearest Lake</li>
<li>Distance to nearest Lake Monitoring Station</li>
</ul>
</li>
</ul>
<p><strong>Potential questions for consideration:</strong></p>
<ul>
<li>Property that is contiguous to a lake, i.e. lake front property, is likely to influence its value. What is the relationship between proximity to water and property value? What effect does location have on this relationship? Does time influence this relationship?</li>
<li>What relationships exist between various property characteristics (i.e. dwelling type, ownership type, etc.) and proximity to water? Have these relationships changed over time? Do the policies / regulations of local or county governments appear to influence these relationships?</li>
<li>Does proximity to water appear to influence how land is being developed over time? Does the amount of influence vary across local or county governments?</li>
</ul>
